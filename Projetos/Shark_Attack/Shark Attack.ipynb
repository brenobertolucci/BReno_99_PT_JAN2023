```python
#importing useful libraries

import pandas as pd
import numpy as np
import re
import openpyxl
```

```python
#importing data

attacks = pd.read_csv("data/attacks.csv", encoding ="latin1",parse_dates = ['Date'], infer_datetime_format = True)
```

```python
attacks["Country"].describe()
```

```python
#Searching for the country with most attack

attacks_by_country = attacks.groupby("Country")["Case Number"].count().sort_values(ascending=False).to_frame(name="Count_Country").reset_index()
attacks_by_country["%"] = attacks_by_country["Count_Country"].transform(lambda x: x/sum(x)) 
```

```python
#Selection only USA
mask_country = attacks["Country"]=="USA"
attack_usa= attacks[mask_country]
```

```python
#Cleaning colunmn Area

list_clean_area = ['Palmyra Atoll','Guam','Johnston Atoll','North & South Carolina',np.nan,'Wake Island','Midway Atoll','East coast','CUBA']
attack_usa["Area"] = attack_usa["Area"].replace(list_clean_area,"Undefined")
attack_usa["Area"] = attack_usa["Area"].str.strip()
```

```python
#Cleaning colunmn Month

pattern_month="[A-Za-z]{3}"

def list_to_string(list_input):
    for item in list_input:
        return str(list_input[0])
list_clean_month = ["Rep","Fal","Ear","Lat","Mid","Sum","Bet","Som","Cir","Bef","dat",np.nan]
    
    
attack_usa["Month"] = attack_usa["Date"].map(lambda date:re.findall(pattern_month,date))
attack_usa["Month"] = attack_usa["Month"].map(list_to_string)
attack_usa["Month"] = attack_usa["Month"].replace(list_clean_month,"Undefined")
```

```python
#Cleaning colunmn Type

attack_usa["Type"] = attack_usa["Type"].replace("Boating","Boat")
attack_usa["Type"] = attack_usa["Type"].replace("Invalid","Undefined")
attack_usa["Type"] = attack_usa["Type"].replace("Invalid","Undefined")
```

```python
#Cleaning column Sex

attack_usa = attack_usa.rename(columns={"Sex ": "Sex"})
attack_usa["Sex"] = attack_usa["Sex"].replace("M ","M")
attack_usa["Sex"] = attack_usa["Sex"].replace("lli","Undefined")
attack_usa["Sex"] = attack_usa["Sex"].replace(np.nan,"Undefined")
```

```python
#Cleaning column Age

def age_buckets(x):
    if x <0:
        return "Undefined"
    elif x>0 and x<=10:
        return "0-10 years"
    elif x>10 and x<=20:
        return "11-20 years"
    elif x>20 and x<=30:
        return "21-30 years"
    elif x>30 and x<=40:
        return "31-40 years"
    elif x>40 and x<=50:
        return "41-50 years"
    elif x>40 and x<=50:
        return "41-50 years"
    elif x>50 and x<=60:
        return "51-60 years"
    else:
        return "Over 60 years"



list_clean_age = ['',np.nan,'20s','50s','Teen','60s','teen','30s','12 or 13','Teens','30s','8 or 10','\xa0 ','6Â½','mid-30s','23 & 20','7      &    31',"60's",'Elderly','mid-20s','Ca. 33','21 or 26','adult','9 & 12','37, 67, 35, 27,  ? & 27','?    &   14','10 or 12','young']
attack_usa["Age"] = attack_usa["Age"].str.strip()
attack_usa["Age"] = attack_usa["Age"].replace(list_clean_age,-1)
attack_usa["Age_Number"]=attack_usa["Age"].astype("int64")


attack_usa["Age_Buckets"] = attack_usa["Age_Number"].apply(age_buckets)
```

```python
#Cleaning column Fatal

attack_usa = attack_usa.rename(columns={"Fatal (Y/N)": "Fatal"})
attack_usa["Fatal"]=attack_usa["Fatal"].replace(np.nan,"Undefined")
attack_usa["Fatal"]=attack_usa["Fatal"].replace("UNKNOWN","Undefined")

```

```python
#Selecting the relevant variables for the studing

variables_selection = ["Country","Area","Month","Sex","Age_Buckets","Type","Fatal"]

attack_usa = attack_usa[variables_selection]
```

```python
#Grouping attacks by state: count

attack_usa["Attack by Area"] = attack_usa.groupby("Area")["Area"].transform("count")
```

```python
#Grouping attacks by month: count

attack_usa["Attack by Month"] = attack_usa.groupby("Month")["Month"].transform("count")
```

```python
#Grouping attacks by sex: count

attack_usa["Attack by Sex"] = attack_usa.groupby("Sex")["Sex"].transform("count")
```

```python
#Grouping attacks by age_buckets: count

attack_usa["Attack by Age"] = attack_usa.groupby("Age_Buckets")["Age_Buckets"].transform("count")
```

```python
#Grouping attacks by type: count

attack_usa["Attack by Type"] = attack_usa.groupby("Type")["Type"].transform("count")
```

```python
#Grouping attacks by fatal: count

attack_usa["Attack by Fatal"] = attack_usa.groupby("Fatal")["Fatal"].transform("count")
```

```python
area_analysis = attack_usa[["Area","Attack by Area"]].sort_values("Attack by Area",ascending = False).drop_duplicates(subset = "Area")
area_analysis["%"] = area_analysis["Attack by Area"].transform(lambda perc: perc/sum(perc))
area_analysis
```

```python
month_analysis = attack_usa[["Month","Attack by Month"]].sort_values("Attack by Month",ascending = False).drop_duplicates(subset = "Month")
month_analysis["%"] = month_analysis["Attack by Month"].transform(lambda perc: perc/sum(perc))
month_analysis
```

```python
sex_analysis = attack_usa[["Sex","Attack by Sex"]].sort_values("Attack by Sex",ascending = False).drop_duplicates(subset = "Sex")
sex_analysis["%"] = sex_analysis["Attack by Sex"].transform(lambda perc: perc/sum(perc))
sex_analysis
```

```python
age_analysis = attack_usa[["Age_Buckets","Attack by Age"]].sort_values("Attack by Age",ascending = False).drop_duplicates(subset = "Age_Buckets")
age_analysis["%"] = age_analysis["Attack by Age"].transform(lambda perc: perc/sum(perc))
age_analysis
```

```python
type_analysis = attack_usa[["Type","Attack by Type"]].sort_values("Attack by Type",ascending = False).drop_duplicates(subset = "Type")
type_analysis["%"] = type_analysis["Attack by Type"].transform(lambda perc: perc/sum(perc))
type_analysis
```

```python
fatal_analysis = attack_usa[["Fatal","Attack by Fatal"]].sort_values("Attack by Fatal",ascending = False).drop_duplicates(subset = "Fatal")
fatal_analysis["%"] = fatal_analysis["Attack by Fatal"].transform(lambda perc: perc/sum(perc))
fatal_analysis
```

```python
#Creating Pivot Table

area_by_month = pd.pivot_table(
    attack_usa[["Area","Month","Sex"]],
    index="Area", columns="Month", 
    aggfunc='count',
    fill_value=0)


area_by_month=area_by_month[[('Sex',       'Jan'),
                             ('Sex',       'Feb'),
                             ('Sex',       'Mar'),
                             ('Sex',       'Apr'),
                             ('Sex',       'May'),
                             ('Sex',       'Jun'),
                             ('Sex',       'Jul'),
                             ('Sex',       'Aug'),
                             ('Sex',       'Sep'),
                             ('Sex',       'Oct'),
                             ('Sex',       'Nov'),
                             ('Sex',       'Dec'),
                             ('Sex',       'Undefined')]]

area_by_month[('Sex',       'Total')] = area_by_month.apply(sum,axis=1)
area_by_month.loc['Total']= area_by_month.sum()
area_by_month = area_by_month.sort_values(by=('Sex',       'Total'),ascending = False)
```

```python
#Exporting to excel tables used in the analysis

with pd.ExcelWriter('data\Shark_Attack.xlsx') as writer:
    attacks_by_country.to_excel(writer, sheet_name='Attacks_by_Country')
    attack_usa.to_excel(writer, sheet_name='Attacks_USA')
    area_analysis.to_excel(writer, sheet_name='Area_Analysis')
    month_analysis.to_excel(writer, sheet_name='Month_Analyis')
    area_by_month.to_excel(writer, sheet_name='Area_by_Month_Analyis')
    sex_analysis.to_excel(writer, sheet_name='Sex_Analyis')
    age_analysis.to_excel(writer, sheet_name='Age_Analyis')
    type_analysis.to_excel(writer, sheet_name='Type_Analyis')
    fatal_analysis.to_excel(writer, sheet_name='Fatal_Analyis')
    
```
